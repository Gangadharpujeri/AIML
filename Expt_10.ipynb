```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df=pd.read_csv('auto_mpg.csv')
sns.boxplot(df['horsepower'])
plt.show()
```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce') #coerce means replces non numeric with "?" NAN
q1=df['horsepower'].quantile(0.25)
q3=df['horsepower'].quantile(0.75)
iqr=q3-q1
lower_limit=q1-(1.5*iqr)
upper_limit=q3+(1.5*iqr)


df[(df['horsepower']>upper_limit)|(df['horsepower']<lower_limit)]
upper_cap=df['horsepower'].quantile(0.9)
lower_cap=df['horsepower'].quantile(0.1)
df['horsepower'] = np.where(
    df['horsepower'] > upper_limit, upper_cap,)
np.where(df['horsepower'] < lower_limit, lower_cap, df['horsepower'])
sns.boxplot(df['horsepower'])
```
