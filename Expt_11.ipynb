```python
import pandas as pd
df=pd.read_csv('auto_mpg.csv', usecols=['mpg','horsepower','weight'])
print(df)
```

```python
df['std_mpg'] = (df['mpg'] - df['mpg'].mean()) / df['mpg'].std()
df['horsepower']=pd.to_numeric(df['horsepower'],errors='coerce')
df['std_horsepower'] =(df['horsepower'] - df['horsepower'].mean()) / df['horsepower'].std()
df['weight_mpg'] = (df['weight'] - df['weight'].mean()) / df['weight'].std()
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['mpg_std', 'horsepower_std', 'weight_std']] = scaler.fit_transform(df[['mpg', 'horsepower', 'weight']])
df[['mpg', 'mpg_std', 'horsepower', 'horsepower_std', 'weight', 'weight_std']]
```

```python
from sklearn.preprocessing import OneHotEncoder
df1=pd.read_csv('auto_mpg.csv')
pd.get_dummies(df1, columns=['origin'])
encoder = OneHotEncoder(sparse_output=False)
data=df1[['origin']]
encoded = encoder.fit_transform(data)
print(encoded)
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['origin']))
df1 = pd.concat([df1, encoded_df], axis=1)
df1
```
