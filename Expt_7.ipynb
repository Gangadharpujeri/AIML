```python
import pandas as pd   
import matplotlib.pyplot as plt       
df = pd.read_csv('auto_mpg.csv')
df.head()
```

```python
df['city_milage'] = df['mpg'] * 0.70
df.head()
```

```python
mpg_list = [df['mpg'], df['city_milage']]
plt.boxplot(mpg_list)
plt.title('MPG VS City Mlage')
plt.ylabel('MPG')
plt.xticks([1,2]), ['Average MPG', 'City MPG']
plt.show()
```

```python
import seaborn as sns # type: ignore
sns.scatterplot(data=df, x = 'horsepower', y = 'mpg')
plt.title('Correction between horse power and mpg')
plt.xlabel('Horsepower')
plt.ylabel('MPG')

correction = df['horsepower'].corr(df['mpg'])
print("correction", correction)
```

```python
year_counts = df['model_year'].value_counts().sort_index()
year_counts . plot (kind = "bar")
plt.title('number of cars released each year')
plt.xlabel('model year')
plt.ylabel ('number of cars')
plt.show()
```

```python
import seaborn as sns
sns.histplot (df ['horsepower'], kde=True, bins=20,
color = 'orange')
plt.title('Distribution of horsepower')
plt.xlabel (' Horsepower')
plt.ylabel (' Number of cars')
plt. grid (True)
plt. show()
```
